```python
!pip install qrcode[pil]
import qrcode
import random
import string
import pandas as pd
from PIL import Image

# دالة لتوليد رقم تسلسلي حقيقي
def generate_real_serial_number():
    # توليد رقم تسلسلي حقيقي (مثال: 10 أرقام حقيقية)
    serial_number = ''.join(random.choices(string.digits, k=10))  # 10 أرقام حقيقية
    return serial_number

# دالة لتوليد رقم تسلسلي مزيف
def generate_fake_serial_number():
    # توليد رقم تسلسلي مزيف (مثال: 10 أرقام عشوائية)
    serial_number = ''.join(random.choices(string.digits, k=10))  # 10 أرقام مزيفة
    return serial_number

# دالة لتوليد QR code من رقم تسلسلي
def generate_qr_code(serial_number, is_real):
    qr = qrcode.QRCode(version=1, error_correction=qrcode.constants.ERROR_CORRECT_L, box_size=10, border=4)
    qr.add_data(serial_number)
    qr.make(fit=True)

    img = qr.make_image(fill='black', back_color='white')
    img_path = f'qr_code_{serial_number}.png'
    img.save(img_path)
    return img_path

# توليد 1000 رقم تسلسلي حقيقي
def generate_real_data():
    real_data = []
    for _ in range(1000):  # توليد 1000 رقم حقيقي
        real_serial = generate_real_serial_number()
        real_qr_path = generate_qr_code(real_serial, True)
        real_data.append({"serial_number": real_serial, "qr_code_path": real_qr_path, "is_real": True})
    return real_data

# توليد 1000 رقم تسلسلي مزيف
def generate_fake_data():
    fake_data = []
    for _ in range(1000):  # توليد 1000 رقم مزيف
        fake_serial = generate_fake_serial_number()
        fake_qr_path = generate_qr_code(fake_serial, False)
        fake_data.append({"serial_number": fake_serial, "qr_code_path": fake_qr_path, "is_real": False})
    return fake_data

# فصل توليد الأرقام الحقيقية والمزيفة
real_data = generate_real_data()
fake_data = generate_fake_data()

# دمج البيانات الحقيقية والمزيفة في قاعدة بيانات واحدة
all_data = real_data + fake_data

# تحويل البيانات إلى DataFrame
df = pd.DataFrame(all_data)

# حفظ قاعدة البيانات في ملف CSV
df.to_csv('serial_numbers_with_qr_codes_separated.csv', index=False)

# عرض بعض البيانات للتأكد
print(df.head())
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
```

```python
# تحميل البيانات من CSV (يجب أن يكون لديك الأعمدة المناسبة مثل الرقم التسلسلي وQR code path)
df = pd.read_csv('serial_numbers_with_qr_codes_separated.csv')

# هنا تحتاج لتحميل الصور من مسارات QR code وحولها إلى numpy arrays
# سنفترض أنك خزنت مسارات الصور في عمود 'qr_code_path'

def load_images(qr_code_paths):
    images = []
    for path in qr_code_paths:
        img = image.load_img(path, target_size=(64, 64))  # Resize الصورة إلى 64x64
        img_array = image.img_to_array(img) / 255.0  # تحويل الصورة إلى مصفوفة وتطبيعها
        images.append(img_array)
    return np.array(images)

# تحميل صور QR code
qr_images = load_images(df['qr_code_path'])
```

```python
import numpy as np

# مثال على دالة encoding تقوم بتحويل الرقم التسلسلي إلى تمثيل ثابت
def encode_serial_number(serial):
    # على سبيل المثال، يمكن تحويل الرقم التسلسلي إلى تمثيل عددي
    return np.array([int(digit) for digit in str(serial)])

# البيانات الوهمية
df = pd.DataFrame({
    'serial_number': ['1234567890', '0987654321', '5678901234']  # أرقام تسلسلية كمثال
})

# تطبيق الدالة على الأرقام التسلسلية
serial_numbers = np.array([encode_serial_number(serial) for serial in df['serial_number']])

# تأكد من أن جميع الأرقام التسلسلية تم تمثيلها على شكل مصفوفة أحادية البُعد
print(serial_numbers)
```

```python
# تقسيم البيانات إلى بيانات تدريب وتحقق

X_train, X_val, y_train, y_val = train_test_split(qr_images, df['is_real'], test_size=0.2, random_state=42)
```

```python
# بناء نموذج CNN بسيط
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')  # استخدم sigmoid للتصنيف بين حقيقي ومزيف
])

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# تدريب النموذج
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val))
```

```python
# تقييم النموذج
val_loss, val_accuracy = model.evaluate(X_val, y_val)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")
```
