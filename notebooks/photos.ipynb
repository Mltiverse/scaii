```python
# ğŸ“Œ 1ï¸âƒ£ Ø§Ø³ØªÙŠØ±Ø§Ø¯ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import numpy as np

# ğŸ“Œ 2ï¸âƒ£ ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
train_dir = "/content/drive/MyDrive/dataset/train"
test_dir = "/content/drive/MyDrive/dataset/test"

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=60,  # ğŸ”¹ Ø²ÙŠØ§Ø¯Ø© Ø§Ù„ØªØ¯ÙˆÙŠØ± Ø­ØªÙ‰ 60 Ø¯Ø±Ø¬Ø©
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.4,
    brightness_range=(0.5, 1.5),  # ğŸ”¹ ØªÙˆØ³ÙŠØ¹ Ù†Ø·Ø§Ù‚ Ø§Ù„Ø³Ø·ÙˆØ¹ Ø£ÙƒØ«Ø±
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode="nearest",
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input  # ğŸ”¹ ØªØ­Ø³ÙŠÙ† Ø§Ù„ØªÙˆØ²ÙŠØ¹ Ø§Ù„Ù„ÙˆÙ†ÙŠ
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(128, 128), batch_size=32, class_mode='binary', subset='training')

val_generator = train_datagen.flow_from_directory(
    train_dir, target_size=(128, 128), batch_size=32, class_mode='binary', subset='validation')

test_generator = test_datagen.flow_from_directory(
    test_dir, target_size=(128, 128), batch_size=32, class_mode='binary')

# ğŸ“Œ 3ï¸âƒ£ Ø­Ø³Ø§Ø¨ class weights Ù„Ù…Ø¹Ø§Ù„Ø¬Ø© Ø¹Ø¯Ù… Ø§Ù„ØªÙˆØ§Ø²Ù†
class_counts = train_generator.classes
class_weights = {0: len(class_counts) / np.count_nonzero(class_counts == 0),
                 1: len(class_counts) / np.count_nonzero(class_counts == 1)}

# ğŸ“Œ 4ï¸âƒ£ ØªØ­Ù…ÙŠÙ„ MobileNetV2 Ø¨Ø¯ÙˆÙ† Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ø¹Ù„ÙˆÙŠØ©
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
base_model.trainable = True  # ğŸ”¹ Ø¥ØªØ§Ø­Ø© Fine-Tuning

# ğŸ“Œ 5ï¸âƒ£ ÙÙŠ Ø§Ù„Ø¨Ø¯Ø§ÙŠØ©ØŒ Ù†Ø¬Ù…Ø¯ ÙƒÙ„ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª
for layer in base_model.layers:
    layer.trainable = False

# ğŸ“Œ 6ï¸âƒ£ Ø¥Ø¶Ø§ÙØ© Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ø¬Ø¯ÙŠØ¯Ø©
x = base_model.output
x = Conv2D(256, (3,3), activation='relu', padding='same')(x)  # ğŸ”¹ Ø±ÙØ¹ Ø¹Ø¯Ø¯ Ø§Ù„ÙÙ„Ø§ØªØ±
x = BatchNormalization()(x)
x = MaxPooling2D(2,2)(x)

x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)
output = Dense(1, activation='sigmoid')(x)

# ğŸ“Œ 7ï¸âƒ£ Ø¥Ù†Ø´Ø§Ø¡ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model = Model(inputs=base_model.input, outputs=output)

# ğŸ“Œ 8ï¸âƒ£ Ø¶Ø¨Ø· Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=1e-4),  # ğŸ”¹ ØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù… Ø§Ù„Ø£Ø³Ø§Ø³ÙŠ
    metrics=['accuracy']
)

# ğŸ“Œ 9ï¸âƒ£ Ø¥Ù†Ø´Ø§Ø¡ Callbacks Ù„ØªØ­Ø³ÙŠÙ† Ø§Ù„ØªØ¯Ø±ÙŠØ¨
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)  # ğŸ”¹ Ø¶Ø¨Ø· ØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù…
]

# ğŸ“Œ ğŸ”Ÿ **Ø§Ù„Ù…Ø±Ø­Ù„Ø© Ø§Ù„Ø£ÙˆÙ„Ù‰: ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ø¹Ù„ÙŠØ§ ÙÙ‚Ø·**
epochs = 10  # ğŸ”¹ ØªØ¯Ø±ÙŠØ¨ Ø³Ø±ÙŠØ¹ ÙÙŠ Ø§Ù„Ø¨Ø¯Ø§ÙŠØ©
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# ğŸ“Œ 1ï¸âƒ£1ï¸âƒ£ **Ø§Ù„Ù…Ø±Ø­Ù„Ø© Ø§Ù„Ø«Ø§Ù†ÙŠØ©: ÙØªØ­ Ø¬Ù…ÙŠØ¹ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ù„Ù„ØªØ¯Ø±ÙŠØ¨**
for layer in base_model.layers:
    layer.trainable = True  # ğŸ”¹ ÙØªØ­ ÙƒÙ„ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª

# ğŸ”¹ ØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù… Ø¨Ø¹Ø¯ Ø§Ù„ÙØªØ­
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=1e-6),  # ğŸ”¹ ØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù… Ø¬Ø¯Ù‹Ø§
    metrics=['accuracy']
)

# ğŸ“Œ 1ï¸âƒ£2ï¸âƒ£ ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ø¹Ø¯ ÙØªØ­ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª
epochs = 30  # ğŸ”¹ Ù…ÙˆØ§ØµÙ„Ø© Ø§Ù„ØªØ¯Ø±ÙŠØ¨ Ø¨Ø¹Ø¯ Ø§Ù„ÙØªØ­
history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# ğŸ“Œ 1ï¸âƒ£3ï¸âƒ£ Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¹Ù„Ù‰ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±
test_loss, test_acc = model.evaluate(test_generator)
print(f"âœ… Ø¯Ù‚Ø© Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¹Ù„Ù‰ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {test_acc:.2%}")

# ğŸ“Œ 1ï¸âƒ£4ï¸âƒ£ Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.save("mobilenet_finetuned_model_v4.keras")
print("âœ… ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ù†Ø¬Ø§Ø­!")
```
